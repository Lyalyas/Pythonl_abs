```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import scipy 

def pendulum(t, s): 
    theta, omega = s
    return [omega, -np.sin(theta)]


# # Фазовые портреты
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlabel(r'$\theta$')
ax.set_ylabel(r'$\omega$')
ax.set_xlim([0, 1])
ax.set_ylim([1.5, 2.5])
ax.set_title('Фазовые портреты маятника')
for i in range(100):
    theta0 = np.random.rand() # начальное условие
    omega0 = 1.5 + np.random.rand()
    sol = solve_ivp(pendulum, [0, 30], [theta0, omega0], max_step=0.1)
    print(sol)
    # f(t, y) - правая часть системы( pendulum )
    # затем идёт интервал для интегрирования ( временной интервал )
    # дальше передаются начальные улсовия
    # шаг, используемый для расчёта значения функции методом Рунге-Кутты 4 порядка ( Метод Эйлера - Ренге- Кутт 1го порядка)
    ax.plot(sol.y[0], sol.y[1], 'b-', alpha=0.15)
    #alpha - tranparency

# Зависимость периода от начальной скорости
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlabel(r'$\omega$')
ax.set_ylabel('Период')
ax.set_xlim([0, 2])
ax.set_title('Зависимость периода маятника от начальной скорости')
omega_arr = np.linspace(.1, 2, 50) # задаются начальные скорости
period_arr = []
for omega0 in omega_arr:
    sol = solve_ivp(pendulum, [0, 50], [0.1, omega0]) # статичному углу отклонения каждый раз соответствует новая скорость
    time = sol.t
    theta = sol.y[0]
    periodic_time = time[np.argmax(theta > 0)] - time[np.argmax(theta < 0)]
    #Выводит время между периодами колебаний:
    # Что же означает: theta > 0 ? - Вместо исходного размера будет выведен булевый массив того же размера т.к. он отсортирован изначально
    #(theta имеет сначала подмассив положителных значений (маятник движется в одну сторону), затем отрицательных), 
    # будет выведен первый найденный True или False в зависимости от того, что нужно
    period_arr.append(periodic_time)
ax.plot(omega_arr, period_arr, 'b-')


# Фазовые портреты с solout
theta0_vals = np.random.rand(100) # Задание 100 рандомных услоивий
omega0_vals = 1.5 + np.random.rand(100)
sol = [] # Будет использовано как список точек одной кривой

fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlabel(r'$\theta$')
ax.set_ylabel(r'$\omega$')
ax.set_xlim([0, 1])
ax.set_ylim([1.5, 2.5])
ax.set_title('Фазовые портреты маятника с solout')


def solout(t, y):
    sol.append([t, *y])

for i in range(100):
    intobj = scipy.integrate.ode(pendulum)
    intobj.set_integrator('dopri5') # Ренге-Кутт 5 порядка 
    intobj.set_initial_value([theta0_vals[i], omega0_vals[i]], 0.0) #Задание начальных условий, 0.0 - левый предел интегрирования
    intobj.set_solout(solout) # Указание функции - обработчика 
    sol = [] 
    intobj.integrate(50) # Вызов функции с правым пределом интегрирования 50
    sol_array = np.array(sol)
    ax.plot(sol_array[:,1], sol_array[:,2], 'b-', alpha=0.15)


plt.show()


```

```python
import numpy as np 
from scipy.optimize import minimize
from scipy.interpolate import RectBivariateSpline
import matplotlib.pyplot as plt 
import math
import time

data = np.loadtxt('./Task-9_10/sample_surface.txt', skiprows=1)


x_data = data[:,0]
y_data = data[:,1]
z_data = data[:,4]
max_data, min_data = np.max(z_data), np.min(z_data[ z_data != 0])

x_unique = np.unique(x_data) # значения повторяются, RectBivariateSpline не допускает этого, нужно привести данные к рабочей для него форме  
y_unique = np.unique(y_data)




Z = np.ndarray(shape=(0, 37), dtype=float)

 
for i, x_val in enumerate(x_unique): # распределение z в 2 мерный массив, как если бы составлял meshgrid 
    z_arr = z_data[ x_data == x_val ] 
    # avarage_not_null_value = 0 # Если просто взять и удалить с изображения все нулевые значения
    # #все граничные сплайны будут очень резкими(за 0.00001 шага оказываться на порядок ниже, чем за предыдущие 2.0 шага), добавление промежуточных значения позволит уменьшить резскость
    # avarage_len = 0
    # for i in range(len(z_arr)):
    #     if(z_arr[i] != 0 ): 
    #         avarage_not_null_value += z_arr[i]
    #         avarage_len += 1 
    # avarage_not_null_value /= avarage_len
    # z_arr[ z_arr == 0] = avarage_not_null_value
    Z = np.append( Z, [z_arr], axis=0)

spl2d = RectBivariateSpline(x_unique, y_unique, Z) # spline - интерполяционная функция поверхности

xnew = np.linspace(-10, 370, 36)
ynew = np.linspace(-10, 370, 36)
Xnew, Ynew = np.meshgrid(xnew, ynew)
Znew = spl2d(xnew, ynew) # уточнённые значения интерполяционной функции


for row in Znew: #удаление всех точек, которые стремяться к неопределённым значениям функции
    row[(row < min_data ) | (row > max_data)] = np.nan



def fun(coords, spl):
    x, y = coords
    return spl(x, y)[0, 0] # значения обёрнуты в 2 мерный массив, поэтому нужно раскрыть таким образом 

############################# получение функции spl и jac 5-е задание

def time_of_function(function): # декоратор подсчёта времени
    def wrapped(*args):
        start_time = time.perf_counter_ns()
        res = function(*args)
        print((time.perf_counter_ns() - start_time)/10e9)
        return res
    return wrapped

spl = spl2d.__call__ # явно выразил функцию из сплайна как instance



@time_of_function
def handleComputeMinimum():
    def jac(f):
        df_dx = f.partial_derivative(1, 0) # 1 - первая производная по x, 0 производная по y (сама функция)
        df_dy = f.partial_derivative(0, 1)

        return (df_dx, df_dy)

    (df_dx, df_dy) = jac(spl2d)

    def modGradient(x, y):
        return math.sqrt( df_dx(x, y) ** 2 + df_dy(x, y) ** 2)

    Tk = 1 # шаг

    x_current = 120
    y_current = 10

    eps = .0001

    arr = []
    flag = True

    #метод наискорейшего спуска 

    counter1 = 0


    while flag:
        counter1 += 1
        while spl2d(x_current, y_current) <= spl2d(x_current - Tk * df_dx(x_current, y_current), 
                                y_current - Tk * df_dy(x_current, y_current)):
            Tk = Tk / 2

        if ( modGradient(x_current, y_current ) < eps ):
            flag = False
        # print(x_current, y_current)
        arr.append([x_current, y_current, spl2d(x_current, y_current)[0][0]])
        x_current -= Tk * df_dx(x_current, y_current )[0][0]
        y_current -= Tk * df_dy(x_current, y_current )[0][0]

    funcJac = spl2d(x_current, y_current)
    print(x_current, y_current, funcJac)
    return (x_current, y_current, funcJac, arr)


############################# auto compute minimum
@time_of_function
def minimizeFunc():
    result = minimize(fun, [120, 10], bounds=[(0, 360)], args=(spl) ) # ищет локальный минимум в окрестности точки + 4-е задание
    print(f"minimum (red point ):  x: {result.x[0]} y: {result.x[1]} value:  {result.fun}")
    return result

result = minimizeFunc()
(x_current, y_current, funcJac, arr) = handleComputeMinimum()

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ad = list(zip(*arr))
# ax.scatter(result.x[0], result.x[1], result.fun, c='r', s=200) # При отображении строится не совсем точный график
# print(arr)
ax.scatter(ad[0], ad[1], ad[2], c='r', s=10)
# ax.scatter(x_current, y_current, funcJac, c='r', s=100) # При отображении строится не совсем точный график

ax.plot_surface(Xnew, Ynew, Znew)


plt.show()

```
